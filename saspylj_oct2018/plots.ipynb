```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
```

```python
mpl.rcParams['axes.labelsize']=18
mpl.rcParams['xtick.labelsize']=16
mpl.rcParams['ytick.labelsize']=16
mpl.rcParams['grid.linestyle'] = ''
mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.facecolor'] = 'w'
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['axes.edgecolor'] = 'k'
mpl.rcParams['xtick.bottom'] = True
mpl.rcParams['ytick.left'] = True
mpl.rcParams['legend.fontsize'] = 16
```

```python
data = {'years':[2002, 2003, 2004, 2005, 2006, 2007, 
                 2008, 2009, 2010, 2011, 2012, 2013, 
                 2014, 2015, 2016, 2017, 2018],
        'sas':[1870, 1830, 1980, 1950, 2030, 2230, 
               2400, 2420, 2550, 2630, 3010, 3200, 
               3000, 3140, 3180, 3260, 2270], 
        'sasmd':[146, 161, 200, 273, 221, 267, 320, 
                 311, 375, 442, 447, 602, 507, 536, 
                 640, 644, 495]}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(df['years'], df['sasmd'] / df['sas'], 's',
         markersize=15, color=sns.color_palette('colorblind')[0], 
         label='"small angle scattering" AND "molecular dynamics"')
ax1.set_xlabel('Year')
ax1.set_ylabel('Small angle scattering publications')
ax1.set_xticks([2002, 2006, 2010, 2014, 2018])
#ax1.set_yticks([0, 0.05, .1])
ax1.legend(loc=[0, 0.92], frameon=False)
plt.savefig('./figures/chem_data_py.png', dpi=300)
```
