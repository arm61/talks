```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
```

```python
mpl.rcParams['axes.labelsize']=18
mpl.rcParams['xtick.labelsize']=16
mpl.rcParams['ytick.labelsize']=16
mpl.rcParams['grid.linestyle'] = ''
mpl.rcParams['axes.grid'] = True
mpl.rcParams['axes.facecolor'] = 'w'
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['axes.edgecolor'] = 'k'
mpl.rcParams['xtick.bottom'] = True
mpl.rcParams['ytick.left'] = True
mpl.rcParams['legend.fontsize'] = 16
```

```python
data = {'years':[2002, 2003, 2004, 2005, 2006, 2007, 
                 2008, 2009, 2010, 2011, 2012, 2013, 
                 2014, 2015, 2016, 2017, 2018],
        'sas':[986000, 1050000, 1040000, 1050000, 
                     1060000, 1050000, 1030000, 1050000, 
                     1010000, 874000, 898000, 794000, 
                     586000, 485000, 324000, 194000, 107000], 
        'sasmd':[14500, 18700, 24100, 26300, 
                     31900, 37500, 41400, 48000, 
                     375, 442, 447, 602, 
                     507, 536, 640, 644, 
                     495]}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots(figsize=(10, 7))
ax1.plot(df['years'], df['chem_python'] / df['chemistry'], 's',
         markersize=15, color=sns.color_palette('colorblind')[1], 
         label='chemistry python')
ax1.set_ylim([0, 0.1])
ax1.set_xlabel('Year')
ax1.set_ylabel('Chemistry publications')
ax1.set_xticks([2002, 2006, 2010, 2014, 2018])
ax1.set_yticks([0, 0.05, .1])
ax1.legend(loc=[0, 0.92], frameon=False)
plt.savefig('./figures/chem_data_py.png', dpi=300)
```
